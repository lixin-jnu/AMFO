```python
import folium
import pandas as pd
import numpy as np

data = pd.read_csv('data/site-optus-melbCBD.csv')

_latitude = data.LATITUDE.to_numpy()
_longitude = data.LONGITUDE.to_numpy()
result = np.stack((_latitude, _longitude), axis=1)

sub_1 = np.array([
    30, 89, 52, 57, 93, 53, 1, 56, 12, 116, 38, 50, 86, 76, 84, 28, 22, 98
])
sub_6 = np.array([
    103, 54, 108, 45, 95, 36, 58, 14, 6, 11, 0, 96, 88, 25, 104, 113, 80, 81,
    94, 7, 31, 92, 115, 39, 21, 72, 4, 9, 105, 85, 37, 63, 107, 17, 79, 26
])
mmWave_24 = np.array([
    32, 55, 124, 118, 101, 3, 106, 97, 66, 19, 68, 13, 122, 70, 23, 59, 15, 47,
    10, 87, 102, 65, 29, 2, 73, 5, 75, 114, 117, 77, 33, 35, 60, 8, 46, 111,
    112, 71, 43, 61, 27, 51, 69, 121, 90, 16, 41, 64, 67, 34, 99, 120, 123, 62,
    100, 74, 109, 42, 119, 48, 78, 24, 83, 110, 20, 44, 91, 40, 82, 18, 49
])
```

```python
user = pd.read_csv(r'data/users-melbcbd-generated.csv')

latitude = user.Latitude.to_numpy()
longitude = user.Longitude.to_numpy()
```

```python
map = folium.Map(
    [-37.8143, 144.9632],  #中心点
    tiles="OpenStreetMap",
    attr="OpenStreetMap",
    min_lat=-37.8212,
    max_lat=-37.8073,
    min_lon=144.9512,
    max_lon=144.9713,
    zoom_start=17,
    control_scale=True)

'''
green:1000m;
blue:500m;
red:250m;
'''

for s1 in sub_1:
    bs_green = folium.CustomIcon('imgs/BS-green.png', icon_size=(20, 52))
    folium.Marker(location=result[s1],
                  #popup='Latitude: ' + str(result[s1][0]) + '\nLongitude: ' + str(result[s1][1]),
                  tooltip='Sub-1: ' + str(s1),
                  icon=bs_green).add_to(map)
    folium.Circle(location=result[s1],
                  radius=1000,
                  color='green',
                  weight=2,
                  opacity=0.6).add_to(map)

for s6 in sub_6:
    bs_blue = folium.CustomIcon('imgs/BS-blue.png', icon_size=(20, 52))
    folium.Marker(location=result[s6],
                  #popup='Latitude: ' + str(result[s6][0]) + '\nLongitude: ' + str(result[s6][1]),
                  tooltip='Sub-6: ' + str(s6),
                  icon=bs_blue).add_to(map)
    folium.Circle(location=result[s6],
                  radius=500,
                  color='blue',
                  weight=2,
                  opacity=0.6).add_to(map)

for s24 in mmWave_24:
    bs_red = folium.CustomIcon('imgs/BS-red.png', icon_size=(20, 52))
    folium.Marker(location=result[s24],
                  #popup='Latitude: ' + str(result[s24][0]) + '\nLongitude: ' + str(result[s24][1]),
                  tooltip='mmWave-24: ' + str(s24),
                  icon=bs_red).add_to(map)
    folium.Circle(location=result[s24],
                  radius=250,
                  color='red',
                  weight=2,
                  opacity=0.6).add_to(map)

folium.PolyLine(locations=[[-37.8131, 144.9512], [-37.8073, 144.9713],
                           [-37.8154, 144.9750], [-37.8212, 144.9550],
                           [-37.8131, 144.9512]],
                color='black',
                weight=6).add_to(map)

for i in np.arange(len(latitude)):
    if i % 3 == 1:
        #color = 'green'
        icon = folium.CustomIcon('imgs/User-green.png', icon_size=(15, 15))
    elif i % 3 == 2:
        #color = 'blue'
        icon = folium.CustomIcon('imgs/User-blue.png', icon_size=(15, 15))
    else:
        #color = 'red'
        icon = folium.CustomIcon('imgs/User-red.png', icon_size=(15, 15))
    folium.Marker(
        location=(latitude[i], longitude[i]),
        #popup=str(result[s1]),
        #tooltip='Sub-1: ' + str(s1),
        icon=icon).add_to(map)

map.save('html/7.html')
```
